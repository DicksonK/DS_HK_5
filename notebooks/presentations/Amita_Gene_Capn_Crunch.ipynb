```python
#import libraries
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
from statsmodels.formula.api import ols
from sklearn.feature_selection import f_regression
from sklearn.cross_validation import train_test_split
import pylab as pl
import statsmodels.api as sm
from scipy.stats.stats import pearsonr
```

```python
#load dataset
!curl http://vincentarelbundock.github.io/Rdatasets/csv/MASS/UScereal.csv > cereal.csv
```

```python
# Basic metrics
# Additional details about data set found at
# http://vincentarelbundock.github.io/Rdatasets/doc/MASS/UScereal.html
# Note the data has been normalized for a 1 cup portion size
# Data quality was validated using R studio -- it's a small dataset (65 entries) and luckily we could keep them all.
cereal = pd.read_csv('cereal.csv')
cereal.describe()
```

```python
cereal.info()
```

```python
# We thought it would be fascinating to find a predictive model for the mineral contents in a cup of your favorite cereal!
# Decided to estimate the following models:
#
# sodium ~ fat + carbo + protein + sugars + fibre + sugars:carbo + fibre:carbo
# potassium ~ fat + carbo + protein + sugars + fibre + sugars:carbo + fibre:carbo
#
# Rationionalized to drop 'calories' since it's entirely composed of fat, carbs, and protein, so it wouldn't capture 
# any new information. 
# Also rationalized to drop manufacturer, 'mfr', on the basis that we don't know the criteria for cereal selection of
# the manufacturer's range
# 'Shelf' was also dropped - it didn't make intuitive sense that a unknown store's display choices would have a (any?) 
# relationship to the nutrition information
# All except 3 cererals were vitamin enriched -- without concrete details on what 'vitamin enrichment' means, didn't feel
# this field added significant value to our model
```

```python
# Exploring relationships -- is there a potential relationship between response and predictors?
s_x = cereal[['fat', 'carbo', 'protein', 'sugars', 'fibre']]
s_y = cereal['sodium']

# Uses 33% of the data set (test_size) to get training samples
s_x_train, s_x_test, s_y_train, s_y_test = train_test_split(s_x, s_y, test_size = 0.33, random_state=123)

# p-values from univariate regression
p_values = f_regression(s_x_train, s_y_train)
p_values
```

```python
# Same exploration for potassium
p_x = cereal[['fat', 'carbo', 'protein', 'sugars', 'fibre']]
p_y = cereal['potassium']

# Uses 33% of the data set (test_size) to get training samples
p_x_train, p_x_test, p_y_train, p_y_test = train_test_split(p_x, p_y, test_size = 0.33, random_state=123)

# p-values from univariate regression
p_values = f_regression(p_x_train, p_y_train)
p_values
```

```python
# log normalization
cereal['log_sodium'] = np.log(cereal['potassium'].values)
cereal['log_potassium'] = np.log(cereal['potassium'].values)
cereal['log_fat']= np.log(cereal['fat'].values)
cereal['log_carbo'] = np.log(cereal['carbo'].values)
cereal['log_fibre'] = np.log(cereal['fibre'].values)
cereal['log_protein'] = np.log(cereal['protein'].values)
cereal['log_sugars'] = np.log(cereal['sugars'].values)

# In hindsight - many of the data values are 0 (i.e. 0g fat, etc) 
# Log transformation isn't well-defined for those entries - we could have added a small value to all data points
# or used a different transformation. 
```

```python
# reshape the data from a pandas series to columns 
y = np.matrix(cereal['potassium']).transpose()

x1 = np.matrix(cereal['fibre']).transpose()

# put the columns together to create an input matrix 
x = np.column_stack([x1])
p_x = sm.add_constant(x)

# create a linear model and fit it to the data
p_results = sm.OLS(y,p_x).fit()
print p_results.summary()
print 'P-Values: ', p_results.pvalues
```

```python
# Linear model
regr = linear_model.LinearRegression()
regr.fit(y, p_x)
 
# Display the coefficients:
print regr.coef_

# Display SSE:
print np.sum((regr.predict(y) - p_x) ** 2)

# Scoring (closer to 1 is better!)
print regr.score(y, p_x)
```

```python
# Add more 
y = np.matrix(cereal['potassium']).transpose()

x1 = np.matrix(cereal['fibre']).transpose()
x2 = np.matrix(cereal['fat']).transpose()
x3 = np.matrix(cereal['protein']).transpose()
x4 = np.matrix(cereal['carbo']).transpose()
x5 = np.matrix(cereal['sodium']).transpose()

x = np.column_stack([x1, x2, x3, x4, x5])
p_x = sm.add_constant(x)

p_results = sm.OLS(y,p_x).fit()
print p_results.summary()
print 'P-Values: ', p_results.pvalues
```

```python
# Are there correlations?
# Calculates a Pearson correlation coefficient and the p-value for testing non-correlation
# Returns pearson coerf and 2-tailed p value
pearsonr(cereal['fibre'], cereal['carbo'])
```
